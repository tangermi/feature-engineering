```python
import nltk
import string
import re

stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()
```

```python
def clean_text(text):
    text = ''.join([word.lower() for word in text if word not in string.punctuation])
    tokens = re.split('\W+', text)
    text = [ps.stem(word) for word in tokens if ps.stem(word) in top100_list]
    return text
```

```python
top_N = 100
```

```python
text = df['content'].str.lower().str.cat(sep=' ')
text = ''.join([word for word in text if word not in string.punctuation])
tokens = re.split('\W+', text)
tokens_stem = [ps.stem(token) for token in tokens]
word_dist = nltk.FreqDist(tokens_stem)
```

```python
stopwords = nltk.corpus.stopwords.words('english')
words_except_stop_dist = nltk.FreqDist(w for w in tokens_stem if w not in stopwords) 
```

```python
print('All frequencies, NOT including STOPWORDS:')
print('=' * 60)
rslt = pd.DataFrame(words_except_stop_dist.most_common(top_N),
                    columns=['Word', 'Frequency'])
print(rslt)
print('=' * 60)
```

```python
words_except_stop_dist.most_common(top_N)[:5]

top100_list = []
for a, b in words_except_stop_dist.most_common(top_N):
    top100_list.append(a)

top100_list[:5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(analyzer=clean_text)
X_count = count_vect.fit_transform(df['content'])
print(X_count.shape)
print(count_vect.get_feature_names())
```

```python
X_count_df = pd.DataFrame(X_count.toarray())
X_count_df.columns = count_vect.get_feature_names()
X_count_df[-5:]
```

```python
vectorized_df = pd.concat([df['user_id'],df['hashtag'],X_count_df],axis = 1)
vectorized_df[:5]
vectorized_df = vectorized_df.astype({'hashtag':str})
vectorized_df.info()
```

```python
user_to_term = vectorized_df.groupby(['user_id','hashtag'])
```

```python
user_to_term_sum=user_to_term.sum()
user_to_term_nonzeros =user_to_term_sum.loc[(user_to_term_sum!=0).any(axis=1)]
```

## Jaccard similarity of content used in two users contents for a given topic

```python
import sklearn.metrics

def jaccard_similarity_to_topic(df,user1,user2,topic):
    columns_names = list(df.columns)
    x1 = [item for item in columns_names if df.loc[user1,topic][item]!= 0]
    x2 = [item for item in columns_names if df.loc[user2,topic][item]!= 0]
    intersection= len(set(x1)&set(x2))
    union = []
    union.extend(x1)
    union.extend(x2)
    union = len(set(union))
    score =intersection/union
    print('Score: %.2f' % (score))
```

```python
#an example of usage
jaccard_similarity_to_topic(user_to_term_nonzeros,1437,10336,"['ff']")
```
