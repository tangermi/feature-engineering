```python
import numpy as np
import pandas as pd

df = pd.read_csv('user_to_mention.csv',sep=',')
```

```python
df = df.drop(['user_id'],axis=1)
```

```python
columns_names = list(df.columns)
```

```python
columns_names.remove('screen_name')
```

```python
mention_dic = {}
for index,row in df.iterrows():
    screen_name = row['screen_name']
    row[screen_name] = 0
    mention_dic[screen_name] = [item for item in columns_names if row[item]!= 0]
```

```python
mention_dict = {k: v for k, v in mention_dic.items() if len(v)!=0}
```
