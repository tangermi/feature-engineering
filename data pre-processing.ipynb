```python
df = pd.concat([rawdata_polluters_tweets,rawdata_legimaters_tweets])
df[:5]
```

```python
df_with_hashtag = df[df['content'].str.contains('#(\w+)')]
df_with_hashtag[:5]
```

```python
df_with_mention = df[df['content'].str.contains('@(\w+)')]
df_with_mention[:5]
```

## build a dataframe of [user_id,hashtag,mention]

```python
df_3d = df_with_mention
df_3d['mention'] = df_3d['content'].apply(lambda x: re.findall(r"@(\w+)", x))
```

```python
df_3d[:5]
```

### for those tweets with mentions, get the top500 frequent hashtags

```python
df_3d['hashtag'] = df_3d['content'].apply(lambda x: re.findall(r"#(\w+)", x))
```

```python
df_3d[:5]
```

```python
hashtag_to_list = df_3d['hashtag'].tolist()
```

```python
flat_list = []
for sublist in hashtag_to_list:
    for item in sublist:
        flat_list.append(item)
```

```python
flat_list_np = np.array(flat_list)
unique,counts = np.unique(flat_list_np, return_counts=True)
hashtag_rank_np = np.asarray((unique,counts)).T
```

```python
hashtag_rank_pd = pd.DataFrame(hashtag_rank_np)
hashtag_rank_pd[1] = pd.to_numeric(hashtag_rank_pd[1])
hashtag_rank_pd = hashtag_rank_pd.sort_values(by=[1],ascending = False)
hashtag_top500 = hashtag_rank_pd[:500]
hashtag_top500
```

```python
set(df_3d.iloc[100]['hashtag'])& set(hashtag_top500[0]) != set()
```

```python
df_3d['Used Frequent Tags'] = df_3d['hashtag'].apply(lambda x: set(x) & set(hashtag_top500[0]) != set())
```

```python
df_3d = df_3d[df_3d['Used Frequent Tags']==True]
df_3d = df_3d.drop(['Used Frequent Tags'],axis=1)
```
