```python
import numpy as np
import pandas as pd
```

```python
user_list = pd.read_csv(r'id_screenName_list.csv',sep=',')
```

```python
user_id_list = user_list['user_id'].tolist()
```

```python
user_name_list = user_list['screen_name'].tolist()
```

```python
from ast import literal_eval
df = pd.read_csv('export_dataframe2.csv', sep=',',converters={'mention':literal_eval})
```

```python
df = df[['user_id','mention']]
```

```python
df[:5]
```

```python
df.info()
```

## filter out those have mentioned people NOT IN the list

```python
df_mention_inlist = df[df['mention'].apply(lambda a: set(a)&set(user_name_list) != set())]
```

```python
df_mention_inlist.info()
```

```python
df_mention_inlist[-5:]
```

```python
df_mention_inlist_play = df_mention_inlist.copy()
```

```python
for index,row in df_mention_inlist_play.iterrows():
    for mention in row['mention']:
        if mention in user_name_list:
            df_mention_inlist_play.at[index,mention] = 1
```

```python
df_mention_inlist_play = df_mention_inlist_play.fillna(0)
```

```python
To_int_list = list(df_mention_inlist_play.columns)
To_int_list.remove('mention')
#for hashtag in hashtag_list:
#    df_mention_inlist_play.astype('int32')
```

```python
df_mention_inlist_play = df_mention_inlist_play.astype({hashtag:int for hashtag in To_int_list})
```

```python
print(df_mention_inlist_play.shape)
df_mention_inlist_play[:5]
```

```python
#Add screen names to them
name_added_df = df_mention_inlist_play.join(user_list.set_index('user_id'), on='user_id',how='inner')
```

```python
print(name_added_df.shape)
name_added_df[:5]
```

```python
cols = name_added_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
name_added_df = name_added_df[cols]
```

```python
name_added_df[:5]
```

```python
user_to_mention = name_added_df.groupby(['user_id','screen_name'])
```

```python
user_to_mention.sum()
```

```python
user_to_mention.sum().reset_index()
```

```python
export_csv = user_to_mention.sum().reset_index().to_csv (r'user_to_mention.csv', index = None, header=True)
```
